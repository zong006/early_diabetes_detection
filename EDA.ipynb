```python
corr_matrix = df.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix - One-Hot Encoded Variables')
plt.show()
```

```python
target_variable = 'class_Positive'
correlation_threshold = 0.4 

filtered_variables = corr_matrix[corr_matrix[target_variable] > correlation_threshold].index

print(f"Variables with correlation > {correlation_threshold} with '{target_variable}':")
print(filtered_variables)
```

The following variables are more correlated with being a positive case:
- Polyuria_Yes
- Polydipsia_Yes

Nonetheless, correlation score of polyuria and polydipsia are quite high, at 0.6

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Polyuria_Yes'], df['Polydipsia_Yes'])

chi2, p, _, _ = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("P-value:", p)

```

A chi-squared test of independence tells us that the  p-value < 0.05, the null hypthesis that these are independent variables are rejected, and we conclude that there is a significant association between "Polyuria_Yes" and "Polydipsia_Yes". Nonetheless, it is unclear without more information on whether one of them are redudndant or not.
